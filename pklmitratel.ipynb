```python
import numpy as np

# Data historis trouble ticket
historical_data = [647, 560, 357, 242, 276, 119, 303, 278, 216, 224, 357, 381]

# Menghitung rata-rata dari data historis
average_tickets = np.mean(historical_data)

# Prediksi untuk bulan berikutnya
predicted_ticket1 = average_tickets

print("Metode Rata-rata - Prediksi Trouble Ticket untuk Bulan Berikutnya:", predicted_ticket1)
```

```python
import pandas as pd

df = pd.DataFrame({'tickets': historical_data})
# Menghitung Simple Moving Average (SMA)
sma_tickets = df['tickets'].rolling(window=3).mean().iloc[-1]

# Prediksi untuk bulan berikutnya
predicted_ticket2 = sma_tickets

print("Metode Moving Average - Prediksi Trouble Ticket untuk Bulan Berikutnya:", predicted_ticket2)
```

```python
# Data historis trouble ticket
historical_data = [647, 560, 357, 242, 276, 119, 303, 278, 216, 224, 357, 381]

# Menghitung persentase pertumbuhan dari bulan terakhir ke bulan sebelumnya
growth_rate = (historical_data[-1] - historical_data[-2]) / historical_data[-2]

# Prediksi trouble ticket untuk bulan berikutnya
predicted_ticket3 = historical_data[-1] * (1 + growth_rate)

print("Metode Persentase Pertumbuhan - Prediksi Trouble Ticket untuk Bulan Berikutnya:", predicted_ticket3)
```

```python
# Ground truth (nilai sebenarnya) untuk bulan berikutnya
actual_tickets = [647, 560, 357, 242, 276, 119, 303, 278, 216, 224, 357, 381]  # Gantilah dengan nilai sebenarnya jika tersedia

# Metode Rata-rata
accuracy_rata_rata = np.mean(np.abs(np.array(predicted_ticket1) - np.array(actual_tickets)) / np.array(actual_tickets)) * 100

# Metode Moving Average
accuracy_moving_average = np.mean(np.abs(np.array(predicted_ticket2) - np.array(actual_tickets)) / np.array(actual_tickets)) * 100

# Metode Persentase Pertumbuhan
accuracy_growth_percentage = np.mean(np.abs(np.array(predicted_ticket3) - np.array(actual_tickets)) / np.array(actual_tickets)) * 100

print("Akurasi Metode Rata-rata:", round(100 - accuracy_rata_rata), "%")
print("Akurasi Metode Moving Average:", round(100 - accuracy_moving_average), "%")
print("Akurasi Metode Persentase Pertumbuhan:", round(100 - accuracy_growth_percentage), "%")
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error

def align_data_length(actual, predicted):
    min_length = min(len(actual), len(predicted))
    actual = actual[:min_length]
    predicted = predicted[:min_length]
    return actual, predicted

# Contoh penggunaan
actual_values = jumlahperbulan['Trouble Ticket']
predicted_values = [406]

# Menyamakan panjang data
aligned_actual, aligned_predicted = align_data_length(actual_values, predicted_values)

# Menghitung RMSE
rmse_result = np.sqrt(mean_squared_error(aligned_actual, aligned_predicted))
print("Root Mean Squared Error (RMSE):", rmse_result)
```
